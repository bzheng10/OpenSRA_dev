```python
import numpy as np
import os, json, sys
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import rasterio as rio
from rasterio.plot import show
```

```python
import edp.liq as liq
import fcn_gen
```

```python
inputs = pd.read_csv(os.path.dirname(os.path.dirname(os.getcwd())) + '\\CA_Zhu_Full_Inputs.csv')
inputs
```

```python
wtd = inputs['WTD_30m'].values
dr = inputs['DistRiver'].values
dc = inputs['DistCoast'].values
dw = inputs['DistWater'].values
precip = inputs['CA_Precip'].values
vs30 = inputs['WillisVS30'].values
lon = inputs['Longitude'].values
lat = inputs['Latitude'].values
```

```python
find_val = -9999
set_val = np.nan
wtd = fcn_gen.find_set_nan(wtd,find_val,set_val)
dr = fcn_gen.find_set_nan(dr,find_val,set_val)
dc = fcn_gen.find_set_nan(dc,find_val,set_val)
dw = fcn_gen.find_set_nan(dw,find_val,set_val)
precip = fcn_gen.find_set_nan(precip,find_val,set_val)
vs30 = fcn_gen.find_set_nan(vs30,find_val,set_val)
```

```python
dc_cutoff = 20
liq_susc, _, _ = liq.zhu_etal_2017(vs30=vs30, precip=precip, dc=dc, dr=dr, dw=dw, wtd=wtd, dc_cutoff=20)
inputs['liq_susc'] = liq_susc
```

```python
inputs.to_csv(os.path.dirname(os.path.dirname(os.getcwd())) + '\\CA_Zhu_Full_Outputs.csv',)
```

```python
##### State-wide Gas pipelines from John Eidinger
df_pipe = gpd.GeoDataFrame.from_file(os.path.dirname(os.path.dirname(os.getcwd())) + '\\State Gas Pipelines\\California_Natural_Gas_Pipeline.shp')
```

```python
##### CA state border
##### https://data.ca.gov/dataset/ca-geographic-boundaries/resource/3db1e426-fb51-44f5-82d5-a54d7c6e188b
##### https://hub.arcgis.com/datasets/febda4ff242f4f699dc40183833ea141
# df_state_border = gpd.GeoDataFrame.from_file(os.getcwd()+'\\ca-state-boundary\\CA_State_TIGER2016.shp')
df_state_border = gpd.GeoDataFrame.from_file(os.path.dirname(os.path.dirname(os.getcwd())) + '\\California_State_Boundary\\California_State_Boundary.shp')
```

```python
colors_liq_susc = {'very low': np.divide([255,255,255],255),
                   'low': np.divide([236,239,108],255),
                   'moderate': np.divide([255,255, 62],255),
                   'high': np.divide([255,188, 65],255),
                   'very high': np.divide([245,125, 65],255)}
groups = inputs.groupby('liq_susc')
```

```python
nx = 1
ny = 1
figsize = (16,8)
fig, ax = plt.subplots(nx, ny, figsize=figsize)
fontsize_label = 16
fontsize_axis = 12
fontname = 'arial'
##
df_pipe.plot(ax=ax, ec=[0.4,0.4,0.4], alpha=0.4)
df_state_border.plot(ax=ax, ec='k', alpha=0.4)
for key in colors_liq_susc.keys():
    print(key)
    ax.plot(inputs['Longitude'][groups.groups[key]],
            inputs['Latitude'][groups.groups[key]],
            marker='o',label=key,
            linestyle='',linewidth=0,
            ms=2,mfc=colors_liq_susc[key],mec=colors_liq_susc[key])
# for name, group in groups:
#     ax.plot(group['Longitude'],group['Latitude'],marker='o',label=name,linestyle='',ms=4,linewidth=0,mfc=colors_liq_susc[name],mec=colors_liq_susc[name])
plt.legend()
ax.set_xlim([-125,-114])
ax.set_xlim([-122.5,-121.9])
ax.set_ylim([37.4,38.1])
##
plt.show()
```
